```python
#Exp7: DBScan for outlier detection 
#PART A: Applying on a toy dataset 
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from numpy import random,where
import matplotlib.pyplot as plt

```

```python
random.seed(8)
x,y=make_blobs(n_samples=200,centers=1, cluster_std=.3)
```

```python
plt.scatter(x[:, 0], x[:, 1])
```

```python
dbscan=DBSCAN(eps=0.28,min_samples=20)
```

```python
pred=dbscan.fit_predict(x)
```

```python
pred #-1 implies it is an outlier since we had only 1 cluster, 0 means it is a part of the cluster
```

```python
anomaly_index=where(pred==-1) #indexes of anomaly
```

```python
values=x[anomaly_index]
```

```python
plt.scatter(x[:, 0], x[:, 1])
plt.scatter(values[:,0],values[:,1],color='r')
```

```python
# PART B : dataset having a particular irregular shape 
from sklearn.datasets import make_circles
from sklearn.preprocessing import StandardScaler
```

```python
X,y=make_circles(n_samples=500,random_state=21, factor=0.4,noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
dbscan=DBSCAN(eps=0.3, min_samples=10)
y_pred=dbscan.fit_predict(X)
```

```python
y_pred #-1 are outliers again
```

```python
anomaly_index=where(y_pred==-1) #indexes of anomaly
```

```python
values=X[anomaly_index]
```

```python
plt.scatter(X[:, 0], X[:, 1],c=y)
plt.scatter(values[:,0],values[:,1],color='r')
```

```python
#PART c : applying db scan on a dataset 
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
```

```python
dbscan.fit(iris.data)
```

```python
dbscan.labels_#outliers have labels=-1
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2).fit(iris.data)
```

```python
pca_2d=pca.transform(iris.data)
```

```python
for i in range(0,pca_2d.shape[0]):
  if dbscan.labels_[i]==0:
    c1=plt.scatter(pca_2d[i,0],pca_2d[i,1],c='r',marker='+')
  elif dbscan.labels_[i]==1:
    c2=plt.scatter(pca_2d[i,0],pca_2d[i,1],c='b',marker='*')
  else:
    outlier=plt.scatter(pca_2d[i,0],pca_2d[i,1],c='g',marker='o')
plt.legend([c1,c2,outlier],['cluster1','cluster2','outlier'])
```
