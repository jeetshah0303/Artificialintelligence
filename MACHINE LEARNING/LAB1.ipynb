```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # plotting
```

```python
data=pd.read_csv('test_data9.csv')
data.head()
```

```python
#task2
nRow, nCol = data.shape
print(f'There are {nRow} rows and {nCol} columns')
print("shape of the dataset is",data.shape)
print("the types of dataset is")
print(data.dtypes)
```

```python
#task3
#as you can see there are many null values present in the dataset
data.info()
data.isnull().sum()
```

```python
#replacing null data only in the given columns for demonstration and filling it with relevant data
data['ParentId']=data['ParentId'].fillna(-1)
data['SomeString']=data['SomeString'].fillna('ABCDEF')
data['SomeUrl']=data['SomeUrl'].fillna('www.nmims.edu')
data['SomeEmail']=data['SomeEmail'].fillna('jeet.shah90@nmims.edu.in')
data['SomeInt']=data['SomeInt'].fillna(1010101)
data.head()
```

```python
#task 4
#removing null values from the dataset 
data=data.dropna(subset=['UUID'])
data.shape
#as you can see the number of rows have been reduced since all rows without a UUID have been removed
```
